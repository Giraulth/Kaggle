```python
import pandas as pd
import os
from statistics import mean
```

```python
PATH_SUBMISSION = 'gender_submission.csv'
PATH_TEST = 'test.csv'
PATH_TRAIN = 'train.csv'
```

```python
def preprocessing(df):
    df = pd.get_dummies(df, columns=['Sex'])
    df['Fare_range'] = df['Fare'].apply(lambda x : get_fare_range(x))
    has_cabine = [0 if pd.isnull(x) else 1 for x in df['Cabin']]
    df['Cabin'] = pd.Series(has_cabine)
    df = df.drop(columns=['Name', 'Ticket'])
    mean_age = df.quantile()['Age']
    is_children = [1 if x < 18 else 0 for x in df['Age']]
    df['Children'] = pd.Series(is_children)
    df['Age'] = df['Age'].fillna(mean_age)
    df = pd.get_dummies(df, columns=['Embarked'])
    df = df.rename(columns={'Embarked_C':'Cherbourg'})
    df = df.rename(columns={'Embarked_Q':'Queenstown'})
    df = df.rename(columns={'Embarked_S':'Southampton'})
    df['Fare'] = df['Fare'].fillna(0)
    return df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
 

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
 
y_pred = rfc.predict(X_test)

print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```
