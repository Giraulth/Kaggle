```python
def preprocessing(df):
    df = pd.get_dummies(df, columns=['Sex'])
    df['Fare_range'] = df['Fare'].apply(lambda x : get_fare_range(x))
    has_cabine = [0 if pd.isnull(x) else 1 for x in df['Cabin']]
    df['Cabin'] = pd.Series(has_cabine)
    df = df.drop(columns=['Name', 'Ticket'])
    mean_age = df.quantile()['Age']
    is_children = [1 if x < 18 else 0 for x in df['Age']]
    df['Children'] = pd.Series(is_children)
    df['Age'] = df['Age'].fillna(mean_age)
    df = pd.get_dummies(df, columns=['Embarked'])
    df = df.rename(columns={'Embarked_C':'Cherbourg'})
    df = df.rename(columns={'Embarked_Q':'Queenstown'})
    df = df.rename(columns={'Embarked_S':'Southampton'})
    df['Fare'] = df['Fare'].fillna(0)
    return df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
 

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
 
y_pred = rfc.predict(X_test)

print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

```python
df_test = pd.read_csv(PATH_TEST)
df_test = preprocessing(df_test)
df_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier

def train(model, X_train, y_train, X_test, y_test, res_map):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
    res_map[model] = accuracy_score(y_test, y_pred)
```

```python
model_list = [LogisticRegression(), KNeighborsClassifier(),\
              SVC(), SVC(kernel = 'rbf'), GaussianNB(),\
             DecisionTreeClassifier(criterion = 'entropy'),\
             RandomForestClassifier(), AdaBoostClassifier(),\
             QuadraticDiscriminantAnalysis(), MLPClassifier(alpha=1, max_iter=1000)]

res_map = {}

for model in model_list:
    train(model, X_train, y_train, X_test, y_test, res_map)
```

```python
# Random forest classifier seems better
res_map
```

```python
from sklearn.model_selection import GridSearchCV

n_estimators = [100, 300, 500, 800, 1200]
max_depth = [5, 8, 15, 25, 30]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10] 

hyperparameters_list = dict(n_estimators = n_estimators, max_depth = max_depth,  
              min_samples_split = min_samples_split, 
             min_samples_leaf = min_samples_leaf)

grid_forest = GridSearchCV(rfc, hyperparameters_list, cv = 3, verbose = 1, 
                      n_jobs = -1)

best_forest = grid_forest.fit(X_train, y_train)
```

```python
y_pred = best_forest.predict(df_test)

df_test['Survived'] = y_pred
df_test = df_test[['PassengerId', 'Survived']]
df_test.to_csv('submission.csv', index = False)
```
